## Assignment Week 1: “Hello, Graph World”

In this initial beachhead assignment, a simple graph is created using NetworkX in Python. It represents a social network with 10 nodes (labeled with unique numbers but representing people, for example) and 18 edges (connections or relationships).

```python
import networkx.generators.small
import networkx as nx
import matplotlib.pyplot as plt

# Creates a small sample Krackhardt kite graph using the NetworkX built-in generator
g = networkx.generators.small.krackhardt_kite_graph()

# Displays the Krackhardt kite details
print(g)
print("Nodes: ", g.nodes())
print("Edges: ", g.edges())

# Draws the Krackhardt kite graph with Matplotlib
plt.figure(figsize=(8, 6))
nx.draw(g, with_labels = True, node_size = 800, font_size = 12, font_color='white')
plt.title("Krackhardt Kite Graph")
plt.show()
```
