### Introduction

The text corpus I will be examining is Walt Whitman's "Leaves of Grass". This text is in the public domain and can be found on Project Gutenberg via NLTK's library.

After loading the text, I can answer the following questions:

- How many total unique words are in the corpus?

```python
from nltk.corpus import gutenberg

leaves = gutenberg.words('whitman-leaves.txt')

# Lowercase for accuracy and filter out non-alphabetic words
leaves = [word.lower() for word in leaves if word.isalpha()]

# Print the total number of unique words
unique_words = set(leaves)
print(f'Total unique words: {len(unique_words)}')

# Print the first 10 unique words
print('First 10 unique words:', list(unique_words)[:10])
```

- Taking the most common words, how many unique words represent half of the total words in the corpus?

```python
from nltk.probability import FreqDist
from collections import Counter

# Get the total number of words
total_words = len(leaves)

# Get the most common words and their counts
counts = Counter(leaves)
sorted_common = counts.most_common()

# Calculate how many unique words make up half the total word count
running_total = 0
unique_count = 0
for word, count in sorted_common:
    running_total += count
    unique_count += 1
    if running_total >= total_words / 2:
        break

print(f"Total words: {total_words}")
print(f"Number of unique words to reach 50% of total: {unique_count}")
```


- Identify the 200 highest frequency words in this corpus.

- Create a graph that shows the relative frequency of these 200 words.

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

fdist = FreqDist(leaves)
top_200 = fdist.most_common(200)
print(top_200)

plt.figure(figsize=(12, 6))
fdist.plot(200, cumulative=False)
plt.title('Frequency of Top 200 Words in Walt Whitman\'s Leaves of Grass')
plt.xticks([])
plt.xlabel('')
plt.ylabel('Number of Occurrences')
plt.show()
```

- Does the observed relative frequency of these words follow Zipf’s law? Explain.

Zipf's law describes the phenomenon in natural language in which word frequency, when sorted in decreasing order, is inversely proportional to its rank. Basically, the nth word has a frequency proportional to 1/n. 

So in the example above, the most common word 'the' has a frequency of 10113.

The 2nd most common is 'and' with 5334 appearances. 5334/10113 = 0.53, very close to exactly 1/2.

The 3rd most common word 'of' has a frequency of 4265, which is 4265/10113 = 0.42, which is fairly close to 1/3.

And the 4th is 'i' with a frequency of 2933, which is 2933/10113 = 0.29, which is close to 1/4.

This is consistent with Zipf's law.
